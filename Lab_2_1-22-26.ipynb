# 12. Closure Preserve Values

```python
def outer():
    x = 10
    def inner():
        return x
    return inner

f = outer()
print(f())
```

I said it would print 10 and it did

# 13. Decorators Change Behavior

```python
def deco(fn):
    def wrapper ():
        print("before")
        fn()
        print("after")
    return wrapper

@deco
def hello():
    print("hello")

hello()
```

I said 3 lines will print and 3 did

# 15. Duck Typing

```python
class MyRange:
    def __iter__ (self):
        return self
    def __next__ (self):
        raise StopIteration

    for x in MyRange():
        print(x)
```
