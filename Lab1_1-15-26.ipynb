# Reflection questions 5
The integer is immutable, and then x was rebound locally inside the function.

```python
# example 6 
def change(lst):
    lst.append(99)

data = [1, 2]
change(data)
print(data)
```

# Reflection questions 6
The list is mutable, so Python mutated the list, and now the data includes 99

```python
# example 7
def reset_wrong(lst):
    lst = []

def reset_right(lst):
    lst.clear()

data = [1, 2, 3]
reset_wrong(data)
print(data)

reset_right(data)
print(data)
```

# Reflection questions 7
The list is mutable, so Python rebound lst to a new list and then mutated the original list using .clear().

```python
# example 8
def demo(x):
    print(id(x))
    x.append(5)
    print(id(x))

lst = [1, 2] 
print(id(lst))
demo(lst)
```
